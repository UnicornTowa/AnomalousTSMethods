```python
import numpy as np
import pandas as pd
import matrixprofile as mp
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv').to_numpy()
data = (data - data.mean(axis=1).reshape(-1, 1)) / data.std(axis=1).reshape(-1, 1)
```

```python
fft = np.abs(np.fft.rfft(data, axis=1)) / 1488
```

```python
plt.plot(fft[0])
fig = plt.gcf()
fig.set_figwidth(14)
```

```python
mean = np.mean(fft, axis=0)
mean.shape
```

```python
plt.plot(mean)
fig = plt.gcf()
fig.set_figwidth(14)
```

```python
fft_mean_deviation = fft - mean
fft_mean_deviation.shape
```

```python
metric = np.sqrt(np.sum(np.square(fft_mean_deviation), axis=1))
metric
```

```python
sns.violinplot(metric)
```

```python
quantile = np.quantile(metric, 0.99)
quantile
```

```python
np.column_stack((np.round(-np.sort(-metric), 2), np.argsort(-metric)))
```

```python
def plot_sample(sample):
    _, ax = plt.subplots(figsize=(20, 5))
    ax.plot(data[sample], label='True Row')
    ax.set_title(f'True Row, Sample: {sample}')
    plt.legend()
    plt.show()
```
